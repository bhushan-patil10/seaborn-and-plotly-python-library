## Seaborn 
#### It is a Python data visualization library built on top of Matplotlib, providing a high-level interface for creating attractive and informative statistical graphics.



### Charts in Seaborn

#### Categorical Plots


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading the data file


```python
countries_df = pd.read_csv('top_10_countries.csv')
```

```python
countries_df
```

#### Bar Plots

```python
# simple barplot with seaborn

sns.barplot(
    x='Region',
    y='Population',
    hue='Region',  # this tells Seaborn to color by Region
    data=countries_df,
    palette='tab10',
    legend=False, # remove duplicate legend if you don’t want it
)
```

```python
# Removing the confidence intervals

sns.barplot(
    x='Region',
    y='Population',
    hue='Region',  # this tells Seaborn to color by Region
    data=countries_df,
    palette='tab10',
    legend=False,  # remove duplicate legend if you don’t want it
    errorbar=None

)
```

```python
# Horizontal Orientation

sns.barplot(
    y='Region',
    x='Population',
    hue='Region',  # this tells Seaborn to color by Region
    data=countries_df,
    palette='tab10',
    legend=False, # remove duplicate legend if you don’t want it
     orient="h",
    errorbar=None

)
```

```python
# Adding hue

plt.figure(figsize=(5,5))
sns.barplot(
    x='Region',
    y='Population',
    data=countries_df,
    estimator=sum,
    hue= 'Region',
    errorbar=None,
    palette='tab10'
    
)

plt.legend(loc='center')

plt.show()
```

```python
# Using matplotlib and seaborn concurrently can allow you to format your visual

plt.figure(figsize=(10,10))

sns.barplot(x = 'Country', y='Population', data=countries_df, estimator=sum, errorbar=None, hue='Region')

plt.legend(loc='center')

plt.show()
```

```python
# Using subplots to plot on 2 axes

fig, (ax1,ax2) = plt.subplots(figsize=(20,10),nrows=1,ncols=2, tight_layout=True)

sns.barplot(x=countries_df['Region'],y=countries_df['Population'],  errorbar=None, estimator=sum, ax=ax1)
sns.barplot(x=countries_df['Country'],y=countries_df['Population'], errorbar=None, estimator=sum, ax=ax2)

plt.show()
```

### Relational plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime as dt
```

```python
stocks= pd.read_excel('stocks.xlsx')
```

```python
stocks
```

```python
# changed the date into year in a different column

stocks['Year']= stocks['Date'].dt.strftime('%Y')
```

```python
sns.set_theme()
```

```python
# LINE PLOT
```

```python
stocks.head()
```

### Line Plot

```python
plt.figure(figsize=(20, 10))

sns.lineplot(x='Date', y= 'Close', data=stocks, hue='Symbol', linestyle='--', palette=['#707070','#7cbb00', '#3b5998'])
```

```python
# SUB PLOTS

fig, (ax1, ax2, ax3)= plt.subplots(1,3, figsize=(20,5))

sns.lineplot(x='Date', y= 'Close', data=stocks[stocks['Symbol']=='AAPL'], ax= ax1)
sns.lineplot(x='Date', y= 'Close', data=stocks[stocks['Symbol']=='FB'], ax= ax2)
sns.lineplot(x='Date', y= 'Close', data=stocks[stocks['Symbol']=='MSFT'], ax= ax3)
```

### relplot

```python
# relplot allows you to create a Facet Grid based on categorical variables along the rows and columns

sns.relplot(x='Date', y='Close', data=stocks, hue='Symbol', kind='line')

plt.show()
```

```python
sns.relplot(x= 'Date', y='Close', data=stocks, row= 'Symbol', kind= 'line', height=3, aspect= 3.5)

plt.show()
```

```python
sns.relplot(x= 'Date', y='Close', data=stocks, col= 'Symbol', kind= 'line', height=5, aspect= 1)

plt.show()
```

## Distribution Plots

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

### Combining Rug Plot with Scatter Plot

```python


fig=plt.figure(figsize=(5,5))
ax=fig.add_axes([0,0,1,1])

ax.set_ylim(165,235)

sns.rugplot(x='body_mass_g',y='flipper_length_mm', data=penguins)
sns.scatterplot(x='body_mass_g',y='flipper_length_mm', data=penguins)
```

### histplot

```python
# Creating a Hist Plot
sns.histplot(x='body_mass_g', data=penguins)
```

```python
# Adding a hue

sns.histplot(x='body_mass_g', data=penguins, hue='island')
```

```python
# Creating a 2d Hist Plot

sns.histplot(x='body_mass_g', data=penguins, y='island')
```

### KDE Plot

```python
# Creating a KDE Plot

sns.kdeplot(x='body_mass_g', data = penguins)
```

```python
# Combining with Rug Plot

sns.kdeplot(x='body_mass_g', data=penguins)
sns.rugplot(x='body_mass_g', data=penguins)
```

```python
# Creating a 2d KDE Plot

sns.kdeplot(x='body_mass_g', y = 'flipper_length_mm', data = penguins)
```

```python
# Combining with Scatter Plot

sns.kdeplot(x='body_mass_g',y='flipper_length_mm' ,data=penguins, fill=True)
sns.scatterplot(x='body_mass_g',y='flipper_length_mm' ,data=penguins)
```

### ecdfplot

```python
# Creating ECDF Plot on x axis variable

sns.ecdfplot(x='flipper_length_mm', data = penguins)
```

```python
# Adding a hue

sns.ecdfplot(y='flipper_length_mm', data=penguins, hue='island')

plt.grid()
```

### displot

### A displot can create distribution plots across a Facet Grid based on col and row variables

```python
sns.displot(x='flipper_length_mm', data=penguins, col='island', row='sex', kind='kde')
```
